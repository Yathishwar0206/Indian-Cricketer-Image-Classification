```python
import os
import shutil
import random

dataset_path = "data"
IMAGE_EXTENSIONS = (".jpg", ".jpeg", ".png", ".webp")

# Get player folders
players = [
    folder for folder in os.listdir(dataset_path)
    if os.path.isdir(os.path.join(dataset_path, folder))
    and folder not in ["train", "validation", "test"]
]

print(players)

# Create base folders
for split in ["train", "validation", "test"]:
    os.makedirs(os.path.join(dataset_path, split), exist_ok=True)

# Process each player
for player in players:
    player_path = os.path.join(dataset_path, player)

    # âœ… Only image files (ignore folders)
    images = [
        img for img in os.listdir(player_path)
        if img.lower().endswith(IMAGE_EXTENSIONS)
        and os.path.isfile(os.path.join(player_path, img))
    ]

    if len(images) == 0:
        print(f"âš ï¸ No images found for {player}, skipping")
        continue

    random.shuffle(images)

    total = len(images)
    train_size = int(0.7 * total)
    val_size = int(0.2 * total)

    train_files = images[:train_size]
    val_files = images[train_size:train_size + val_size]
    test_files = images[train_size + val_size:]

    # Create player folders
    for split in ["train", "validation", "test"]:
        os.makedirs(os.path.join(dataset_path, split, player), exist_ok=True)

    # Copy files
    for img in train_files:
        shutil.copy(
            os.path.join(player_path, img),
            os.path.join(dataset_path, "train", player, img)
        )

    for img in val_files:
        shutil.copy(
            os.path.join(player_path, img),
            os.path.join(dataset_path, "validation", player, img)
        )

    for img in test_files:
        shutil.copy(
            os.path.join(player_path, img),
            os.path.join(dataset_path, "test", player, img)
        )

    print(f"âœ… {player}: {len(train_files)} train, {len(val_files)} val, {len(test_files)} test")

print("\nğŸ¯ Dataset split completed successfully!")
```

- The above cell is to load the cricketer images splited into **Training**,**Testing**,**Validation**.
