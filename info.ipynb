## Predicting the player info

```python
with open("Players.json","r") as f:
    player_info = json.load(f)

predicted_class = "Virat Kohli"

player = player_info[predicted_class]

text_output = f"""
Player Name : {player['displayName']}
Role : {player['role']}
Born : {player['born']}
Info : {player['info']}
ODI : {player['ODI']}
Test : {player['Test']}
IPL : {player['IPL']}
Man of the Match : {player['Man_of_the_Match']}
Nickname : {player['Nickname']}
"""
print(text_output)
```

## Observation
- Here, I given **Virat Kohli** info for the prediction and it successfully loaded.

## Voice output

```python
# Voice output
import pyttsx3
engine = pyttsx3.init()
engine.say(text_output)
engine.runAndWait()
```

- It is will speak the info of the player
